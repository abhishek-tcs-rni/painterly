## Import it as a module

```python
%matplotlib inline
from painterly import paint

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('images/tomato.jpg', cv2.IMREAD_COLOR)[:,:,::-1]

plt.imshow(img)
plt.title('Original')
plt.show()

painting = paint(img, [8,4,2], T=20, curved=True, f_g=1)

plt.imshow(painting)
plt.title('Painting')
plt.show()
```
